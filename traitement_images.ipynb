```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import seaborn as sns
```

# chargement de l'image

```python
image = ""

try:
    image = Image.open("images/img1.jpg")
except:
    print("Impossible d'ouvrir l'image")

print(image)
```

# Conversion de l'image en tableau numpy

```python
image_np = np.array(image)
image_np.shape
```

## convertir l'image en noir et blanc

```python
image_bw = image.convert('L')
image_bw_np = np.array(image_bw)
image_bw_np.shape
```

```python
image_bw_np
```

## Afficher l'image avec matplotlib

```python
height_img = image_bw_np.shape[0]
height_img
```

```python
wid_img = image_bw_np.shape[1]
wid_img
```

```python
image_bw_np
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# afficher l'image en noire et blanc
axes[0].imshow(image_bw_np, cmap='gray')
axes[0].set_title('Image en noire et blanc')
axes[0].axis('off')
```
